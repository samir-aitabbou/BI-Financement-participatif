```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn import metrics
```

```python
def valeur_manq(data):
    data_numeric = data.select_dtypes(include='number')
    data_numeric.replace(np.NaN, inplace=True)
    imp = SimpleImputer(missing_values=np.NaN, strategy='mean')
    data_imputed = pd.DataFrame(imp.fit_transform(data_numeric))
    return data_imputed
```

```python
adultrain = valeur_manq(data_train)
```

```python
adultest = valeur_manq(data_test)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
X_adultrain = scaler.fit_transform(adultrain)
X_adultest = scaler.fit_transform(adultest)
```

```python
y_train = data_train['state']
y_test = data_train['state']
```

```python
def accuracy_score(lst_classif,lst_classif_names,X,y):
    for clf,name_clf in zip(lst_classif,lst_classif_names):
        scores = cross_val_score(clf, X, y, cv=5)
        print("Accuracy of "+name_clf+" classifier on cross-validation: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
dummycl = DummyClassifier(strategy="most_frequent")
gmb = GaussianNB()
dectree = tree.DecisionTreeClassifier()
rdforest = RandomForestClassifier()
logreg = LogisticRegression(solver="liblinear")
#svc = svm.SVC(gamma='scale')

lst_classif = [dummycl, gmb, dectree, rdforest, logreg]
lst_classif_names = ['Dummy', 'Naive Bayes', 'Decision tree', 'Random Forest', 'Logistic regression']


accuracy_score(lst_classif,lst_classif_names,X_adultrain,y_train)
```
