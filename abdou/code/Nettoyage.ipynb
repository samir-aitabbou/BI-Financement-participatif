```python
#reste Ã  voir si il y'a autant de male et de female dans successul et failed
```

```python
successful_data = data[data['state'] == 'successful']
failed_data = data[data['state'] == 'failed']

# les occurrences de 'male' et 'female' dans chaque catÃ©gorie
male_count_successful = successful_data[successful_data['sex'] == 'male'].shape[0]
female_count_successful = successful_data[successful_data['sex'] == 'female'].shape[0]

male_count_failed = failed_data[failed_data['sex'] == 'male'].shape[0]
female_count_failed = failed_data[failed_data['sex'] == 'female'].shape[0]

print("Successful - Male count:", male_count_successful)
print("Successful - Female count:", female_count_successful)
print("Failed - Male count:", male_count_failed)
print("Failed - Female count:", female_count_failed)
```

```python
# on a presque les memes valeurs donc nous decidons de suprimer l'attribut sex et les valeurs de states differents de successful et failed
```

```python
data = data[data['state'].isin(['successful', 'failed'])]
data = data.drop('sex', axis=1)
```

```python
# Conversion EUR en USD
```

```python
exchange_rates = {'USD': 1, 'EUR': 1.18}  # 1 EUR = 1.10 USD
data['exchange_rate'] = data['currency'].map(exchange_rates)
```

```python
# Conversion 'goal' et 'pledged'  EUR en USD
data['goal_usd'] = data['goal'] * data['exchange_rate']
data['pledged_usd'] = data['pledged'] * data['exchange_rate']
```
